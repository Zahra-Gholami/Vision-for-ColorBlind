```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

Conversion Matrix for **Protanopia**

```python
def simulate_protanopia(image):
    transform_matrix = np.array([
        [0.567, 0.433, 0],
        [0.558, 0.442, 0],
        [0, 0.242, 0.758]
    ])
    data = np.dot(np.array(image), transform_matrix.T)
    data = np.clip(data, 0, 255).astype(np.uint8)
    return Image.fromarray(data)
```

Conversion Matrix for **Deuteranopia**

```python
def simulate_deuteranopia(image):
    transform_matrix = np.array([
        [0.625, 0.375, 0],
        [0.7, 0.3, 0],
        [0, 0.3, 0.7]
    ])
    data = np.dot(np.array(image), transform_matrix.T)
    data = np.clip(data, 0, 255).astype(np.uint8)
    return Image.fromarray(data)
```

Conversion Matrix for **Tritanopia**

```python
def simulate_tritanopia(image):
    # ماتریس تبدیل برای Tritanopia
    transform_matrix = np.array([
        [0.95, 0.05, 0],
        [0, 0.433, 0.567],
        [0, 0.475, 0.525]
    ])
    # اعمال ماتریس تبدیل به تصویر
    data = np.dot(np.array(image), transform_matrix.T)
    data = np.clip(data, 0, 255).astype(np.uint8)
    return Image.fromarray(data)
```

Increase the contrast of green and blue colors

```python
def optimize_for_protanopia(image):
    data = np.array(image)
    data[:, :, 1] = np.clip(data[:, :, 1] * 1.5, 0, 255)  # green
    data[:, :, 2] = np.clip(data[:, :, 2] * 1.2, 0, 255)  # blue
    return Image.fromarray(data)
```

Increase the contrast of red and blue colors

```python
def optimize_for_deuteranopia(image):
    data = np.array(image)
    data[:, :, 0] = np.clip(data[:, :, 0] * 1.5, 0, 255)  # red
    data[:, :, 2] = np.clip(data[:, :, 2] * 1.2, 0, 255)  # blue
    return Image.fromarray(data)
```

Increase the contrast of green and red colors

```python
def optimize_for_tritanopia(image):
    data = np.array(image)
    data[:, :, 0] = np.clip(data[:, :, 0] * 1.5, 0, 255)  # red
    data[:, :, 1] = np.clip(data[:, :, 1] * 1.2, 0, 255)  # green
    return Image.fromarray(data)
```

Upload image

```python
image = Image.open("image.jpeg").convert("RGB")
```

Color blindness simulation

```python
protanopia_image = simulate_protanopia(image)
deuteranopia_image = simulate_deuteranopia(image)
tritanopia_image = simulate_tritanopia(image)
```

Image optimization for any type of color blindness

```python
optimized_protanopia = optimize_for_protanopia(image)
optimized_deuteranopia = optimize_for_deuteranopia(image)
optimized_tritanopia = optimize_for_tritanopia(image)
```
